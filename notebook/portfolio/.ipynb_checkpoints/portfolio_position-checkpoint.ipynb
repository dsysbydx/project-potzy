```python
import sys
sys.path.append('../../')
sys.path.append('../../potzy/')
%run '../header.ipynb'
from morningstarreader import *
import time
```

```python
USDTHB=31.0
TRADE_LOG_FILE='../../data/trade/tradelog_20181103.csv'
## define get position funcdtino from trade 
def get_portfolio_position(trade_log_file,as_of_date=None):
    df =pd.read_csv(trade_log_file)
    df['date']=pd.to_datetime(df['Date'])
    del df['Date']
    if as_of_date:
        df = df[df['date']<=as_of_date]
    #get position
    df['signed_position']=((df['Action']=='B')-0.5)*2.0
    df['signed_position']=df['signed_position']*(df['No.'].apply(lambda x: float(x.replace(',',''))))
    df_pos=df.groupby(['Ticker','Country'])['signed_position'].sum().reset_index() 
    df_pos.columns = [x.lower() for x in df_pos.columns]
    df_pos.rename(columns={'country':'region'},inplace=True)
    return df_pos
```

```python
#load position file
df = get_portfolio_position(TRADE_LOG_FILE,as_of_date=dt.date(2018,4,1))
df.loc[df['ticker']=='DIS','ticker']='XNYS:DIS'
```

```python
kr = KeyRatioData()
kr.load_data('CPALL','th')
```

```python
## loop over all ticker region to laod data
kr = KeyRatioData()
ticker_col_idx=df.columns.get_loc('ticker')
region_col_idx=df.columns.get_loc('region')
for row in df.itertuples():
    time.sleep(1.0)
    kr.load_data(row[ticker_col_idx+1],row[region_col_idx+1])
    print('Done for row:',row)
```

```python
#get eps
eps=np.zeros(shape=(len(df)))
cfps=np.zeros(shape=(len(df)))
for row in df.itertuples():
    eps[row[0]]=float(kr.get_eps(row[ticker_col_idx+1],row[region_col_idx+1]).dropna()[-1])
    cfps[row[0]]=float(kr.get_cfps(row[ticker_col_idx+1],row[region_col_idx+1]).dropna()[-1])
df['eps']=eps
df['cfps']=cfps
df['fx']=1.0
df.loc[df['region']=='us','fx']=USDTHB
df['earning']=df['signed_position']*df['eps']*df['fx']
df['cashflow']=df['signed_position']*df['cfps']*df['fx']
```

```python
df.sort_values(by='cashflow',ascending=False)
```
