```python
import sys
sys.path.append('../../')
sys.path.append('../../potzy/')
%run '../header.ipynb'
from morningstarreader import *
import time
```

```python
USDTHB=31.0
TRADE_LOG_FILE='../../data/trade/tradelog_20181103.csv'
## define get position funcdtino from trade 
def get_portfolio_position(trade_log_file,as_of_date=None):
    df =pd.read_csv(trade_log_file)
    df['date']=pd.to_datetime(df['Date'])
    del df['Date']
    if as_of_date:
        df = df[df['date']<=as_of_date]
    #get position
    df['signed_position']=((df['Action']=='B')-0.5)*2.0
    df['signed_position']=df['signed_position']*(df['No.'].apply(lambda x: float(x.replace(',',''))))
    df_pos=df.groupby(['Ticker','Country'])['signed_position'].sum().reset_index() 
    df_pos.columns = [x.lower() for x in df_pos.columns]
    df_pos.rename(columns={'country':'region'},inplace=True)
    return df_pos
```

```python
#load position file
df = get_portfolio_position(TRADE_LOG_FILE,as_of_date=dt.date(2018,4,1))
df.loc[df['ticker']=='DIS','ticker']='XNYS:DIS'
```

```python
## loop over all ticker region to laod data
kr = KeyRatioData()
ticker_col_idx=df.columns.get_loc('ticker')
region_col_idx=df.columns.get_loc('region')
for row in df.itertuples():
    time.sleep(1.0)
    kr.load_data(row[ticker_col_idx+1],row[region_col_idx+1])
    print('Done for row:',row)
```

```python
#get eps
df_all=pd.DataFrame()
for i in range(5):
    eps=np.zeros(shape=(len(df)))
    cfps=np.zeros(shape=(len(df)))
    for row in df.itertuples():
        eps[row[0]]=float(kr.get_eps(row[ticker_col_idx+1],row[region_col_idx+1]).dropna()[-i-2])
        cfps[row[0]]=float(kr.get_cfps(row[ticker_col_idx+1],row[region_col_idx+1]).dropna()[-i-2])
    df['eps']=eps
    df['cfps']=cfps
    df['fx']=1.0
    df.loc[df['region']=='us','fx']=USDTHB
    df['earning']=df['signed_position']*df['eps']*df['fx']
    df['cashflow']=df['signed_position']*df['cfps']*df['fx']
    df['year_lookback']=2018-i
    df_all = df_all.append(df)
```

```python
df_cf=df_all.pivot(index='ticker',columns='year_lookback',values='cashflow').T
df_cf['All']=df_cf.sum(axis=1)
df_cf=df_cf.T
```

```python
df_cf.sort_values(by=2018,ascending=False)
```

```python
## get price
from pandas_datareader import data
import fix_yahoo_finance as yf
yf.pdr_override() 
```

```python
ticker_col_idx=df.columns.get_loc('ticker')
region_col_idx=df.columns.get_loc('region')
price_dict={}
for row in df.itertuples():
    ticker = row[ticker_col_idx+1]
    region = row[region_col_idx+1]
    
    ticker_mod = ticker.split(':')[-1]
    if region=='th':
        ticker_mod+='.BK'
    
    price_data=data.get_data_yahoo(ticker_mod, start = '2018-04-01',end = '2018-04-28')['Adj Close'].values[-1]
    price_dict[ticker]=price_data
```

```python
df['price']=df['ticker'].apply(lambda x: price_dict[x])
df['equity']=df['price']*df['signed_position']
df['p/cf']=df['price']/df['cfps']
df['p/e']=df['price']/df['eps']
```
