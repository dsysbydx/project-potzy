```python
%run 'header.ipynb'
from pandas_datareader import data
import fix_yahoo_finance as yf
import time
import glob
yf.pdr_override() 

from potzy.metadata import get_all_thai_tickers
all_thai_tickers = get_all_thai_tickers()
print(len(all_thai_tickers))
```

```python
## SET PARAMETERS

#today=dt.datetime.today().date().strftime('%Y%m%d')
start_date = '2000-01-01'
end_date = '2018-03-10' ##end-date is the same as asof for fixing
directory = '../data/price/yahoo/20180310/'
NUM_RETRY = 3

#get all file in there, if none, ensure_dir
all_files = glob.glob(directory+'*.csv')
if len(all_files)==0:
    print('Creating new directory ...')
    ensure_dir(directory)
tickers_done = [x.split('/')[-1][:-4] for x in all_files]
tickers_left = [x for x in all_thai_tickers if x not in tickers_done]
print('Number of tickers in directory:', len(tickers_done))
print('Number of tickers left:', len(tickers_left))
```

```python
### ADD RETRY OPTION
bad_tickers={}
tickers_done, tickers_left = set(tickers_done), set(tickers_left)
num_round=0
while len(tickers_left) > 0 and num_round < NUM_RETRY:
    print('Starting round ',num_round+1)
    print('Number of tickers left are ',len(tickers_left))
    for ticker in tickers_left:
        try:
            symbol = ticker.upper()+'.BK'
            df = data.get_data_yahoo(symbol, start_date, end_date)
            print(ticker,len(df))
            if(len(df) > 1):
                df.to_csv(directory+ticker+'.csv')
                tickers_done.add(ticker)
            time.sleep(2.0)
        except:
            bad_tickers.add(ticker)  
    tickers_left = tickers_left.difference(tickers_done)
    num_round+=1
    time.sleep(5.0)
```
